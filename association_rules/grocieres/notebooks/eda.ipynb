# Análise exploratória de dados

```python
# importando bibliotecas
import pandas as pd
import numpy as np
```

```python
# carregando dataset
df = pd.read_csv("../../../datasets/Groceries_dataset.csv", sep=',', encoding='iso-8859-1')
df
```

```python
# verificando quantidade de linhas e colunas
df.shape
```

```python
# verificando tipos de dados
df.dtypes
```

```python
# verificando valores nulos
df.isnull().sum()
```

```python
# convertendo data para datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df.dtypes
```

```python
# verificando quantidade de vendas por item
df['itemDescription'].value_counts()
```

```python
# verificando quantidade de vendas por item e consumidor
df['Member_number'].value_counts().head(10)
```

```python
# criando dataframe para verificação de vendas por item e consumidor
df2 = df.groupby(['Member_number', 'itemDescription'])['itemDescription'].count()
df2
```

```python
# transformando o dataframe para o formato de tabela
df2 = df2.unstack().fillna(0)
df2
```

```python
# preparando o dataframe para APRIORI
df2 = df2.applymap(lambda x: 1 if x > 0 else 0)
df2
```

```python
# salvando dataframe em csv
df2.to_csv('../data/clean.csv', index=True, header=True, sep=',', encoding='iso-8859-1')
```
