```python
# biblioteca
from pyECLAT import ECLAT, Example1
```

```python
# ingestao de dados
dados = Example1().get()
dados
```

```python
# inicializando eclat
eclat = ECLAT(data=dados, verbose=True)
```

```python
# verificando os itens unicos
eclat.uniq_
```

```python
#
eclat.df_bin
```

```python
# aplicando algoritmo
indices, suporte = eclat.fit(min_support=0.3, min_combination=2, max_combination=4)
```
