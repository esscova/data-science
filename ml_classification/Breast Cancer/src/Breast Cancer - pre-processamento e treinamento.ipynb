## Treinamento de modelos

### bases de treino e teste

```python
# split train test
previsores_dict = {
    'original': X,
    'padronizado' : X_scaled
}
previsores_agrupados={}
for chave, previsor in previsores_dict.items():
    x_train, x_test, y_train, y_test = train_test_split(previsor, y, test_size=0.3, random_state=0)
    previsores_agrupados[f'x_train_{chave}'] = x_train
    previsores_agrupados[f'x_test_{chave}'] = x_test
    previsores_agrupados[f'y_train_{chave}'] = y_train
    previsores_agrupados[f'y_test_{chave}'] = y_test
```

### classificadores

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import roc_auc_score,classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold

import warnings
warnings.filterwarnings("ignore")
```

```python
# função para treinar avaliar um modelo

def treinar_e_avaliar_modelo(clf, nome_modelo, dados, previsores, alvo):
    """
    Treina, avalia o modelo para cada conjunto de previsores e realiza cross-validation.
    Retorna um DataFrame com as principais métricas (acurácia de treino, teste, validação cruzada, etc.).
    """
    resultados = [] 

    for nome, df in previsores.items():
        x_train = dados[f'x_train_{nome}']
        y_train = dados[f'y_train_{nome}']
        x_test = dados[f'x_test_{nome}']
        y_test = dados[f'y_test_{nome}']

        # treino
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)

        # metricas
        acuracia_teste = accuracy_score(y_test, y_pred)
        acuracia_treino = clf.score(x_train, y_train)
        matriz_confusao = confusion_matrix(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)

        # validação cruzada
        cv_scores = cross_val_score(clf, df, alvo, cv=StratifiedKFold(n_splits=30, shuffle=True, random_state=0))
        acuracia_cv = cv_scores.mean() * 100

        # Registro dos resultados
        resultados.append({
            'modelo': nome_modelo,
            'previsores': nome,
            'acuracia_teste': round(acuracia_teste * 100,2),
            'acuracia_treino': round(acuracia_treino * 100,2),
            'acuracia_cv': round(acuracia_cv,2),
            'precision': round(report['weighted avg']['precision'] * 100,2),
            'recall': round(report['weighted avg']['recall'] * 100,2),
            'f1_score': round(report['weighted avg']['f1-score'] * 100,2),
            'matriz_confusao': matriz_confusao.tolist()
        })

    return pd.DataFrame(resultados)
```

#### naive bayes

```python
clf = GaussianNB()
```

```python
naive_model = treinar_e_avaliar_modelo(clf=clf, nome_modelo='Naive Bayes', dados=previsores_agrupados, previsores=previsores_dict, alvo=y.values)
naive_model
```

#### support vector machines

```python
clf_svm = SVC(random_state=0)
```

```python
svm_model = treinar_e_avaliar_modelo(
    clf=clf_svm, 
    nome_modelo='Support Vector Machine', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

svm_model
```

#### logistic regression

```python
clf_lr = LogisticRegression(random_state=0)
```

```python
lr_model = treinar_e_avaliar_modelo(
    clf=clf_lr, 
    nome_modelo='Logistic Regression', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

lr_model
```

#### KNN

```python
clf_knn = KNeighborsClassifier()
```

```python
knn_model = treinar_e_avaliar_modelo(
    clf=clf_knn, 
    nome_modelo='KNN', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

knn_model
```

#### descision tree

```python
clf_tree = DecisionTreeClassifier(random_state=0)
```

```python
tree_model = treinar_e_avaliar_modelo(
    clf=clf_tree, 
    nome_modelo='Árvores de decisão', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

tree_model
```

#### ramdom forest

```python
clf_rf = RandomForestClassifier(random_state=0)
```

```python
rf_model = treinar_e_avaliar_modelo(
    clf=clf_rf, 
    nome_modelo='Random Forest', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

rf_model
```

#### gradient boosting

```python
clf_gb = GradientBoostingClassifier(random_state=0)
```

```python
gb_model = treinar_e_avaliar_modelo(
    clf=clf_gb, 
    nome_modelo='Gradient Boosting', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

gb_model
```

#### xg boosting

```python
clf_xgb = XGBClassifier()
xgb_model = treinar_e_avaliar_modelo(
    clf=clf_xgb, 
    nome_modelo='XGBoosting', 
    dados=previsores_agrupados, 
    previsores=previsores_dict, 
    alvo=y.values)

xgb_model
```

## resultados

```python
resultados = pd.concat([naive_model, svm_model, lr_model, knn_model, tree_model, rf_model,gb_model, xgb_model])
resultados['dif_acuracia_treino_teste'] = resultados['acuracia_treino'] - resultados['acuracia_teste']
resultados
```
